```python
final_rated_books.drop_duplicates(["User-ID","Title"],inplace=True)
```

```python
final_rated_books.head()
```

```python
book_matrix=final_rated_books.pivot_table(columns='User-ID',index='Title',values='Book-Rating')
```

```python
book_matrix.head()
```

```python
book_matrix.fillna(0,inplace=True)
```

```python
book_matrix.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_matrix)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestions=model.kneighbors(book_matrix.iloc[0,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestions
```

```python
for i  in suggestions:
    print(book_matrix.index[i])
```

```python
import pickle
pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(final_rated_books,open('artifacts/final_rated_books.pkl','wb'))
pickle.dump(book_name,open('artifacts/book_name.pkl','wb'))
pickle.dump(book_matrix,open('artifacts/book_matrix.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id=np.where(book_matrix.index==book_name)[0][0]
    distance,suggestions=model.kneighbors(book_matrix.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in suggestions:
        books=book_matrix.index[i]
        for j in books:
            print(j)
    
```

```python
recommend_book('Harry Potter and the Chamber of Secrets (Book 2)')
```
